```python
import random

class Dice:
    def roll(self):
        return random.randint(1, 6)

class Coin:
    def toss(self):
        return random.choice(['Head', 'Tails'])

class Player:
    def __init__(self, name):
        self.name = name
        self.score = 0

    def play_round(self, dice, coin):
        dice_roll = dice.roll()
        if dice_roll % 2 == 0:
            self.score += dice_roll * 2
        else:
            self.score += dice_roll

        coin_toss = coin.toss()
        if coin_toss == 'Head':
            self.score *= 3
        else:
            self.score -= 1

class Game:
    def __init__(self, player1, player2):
        self.player1 = player1
        self.player2 = player2
        self.dice = Dice()
        self.coin = Coin()

    def play(self):
        self.player1.play_round(self.dice, self.coin)
        self.player2.play_round(self.dice, self.coin)

    def get_winner(self):
        if self.player1.score > self.player2.score:
            return self.player1
        else:
            return self.player2

class FinalGame:
    def __init__(self, game1, game2):
        self.game1 = game1
        self.game2 = game2

    def play(self):
        game1_winner = self.game1.get_winner()
        game2_winner = self.game2.get_winner()
        final_game = Game(game1_winner, game2_winner)
        final_game.play()
        return final_game.get_winner()

player1 = Player(input('Player1 Name = '))
player2 = Player(input('Player2 Name = '))
player3 = Player(input('Player3 Name = '))
player4 = Player(input('Player4 Name = '))

# First game 
game1 = Game(player1, player2)
game1.play()

game2 = Game(player3, player4)
game2.play()


final_game = FinalGame(game1, game2)
final_winner = final_game.play()

print("Final winner:", final_winner.name)
```

```python
import random

class Dice:
    def __init__(self):
        self.sides = [1,2,3,4,5,6]
    
    def roll(self):
        return random.choice(self.sides)


class Coin:
    def __init__(self):
        self.choices = ["Head", "Tail"]
    
    def toss(self):
        return random.choice(self.choices)


class Player:
    def __init__(self, name):
        self.name = name
        self.score = 0
    
    def update_score(self, new_score):
        self.score = new_score

        
class Game:
    def __init__(self, player1, player2, dice, coin):
        self.player1 = player1
        self.player2 = player2
        self.dice = dice
        self.coin = coin
    
    def play(self, player):
        print(player.name+"'s turn ..")
        s = self.dice.roll()
        if s%2==0:
            s *= 2
        choice = self.coin.toss()
        if choice=="Head":
            s *= 3
        else:
            s -= 1
        player.update_score(s)
        print(player.name+"'s score: ",player.score)
    
    def select_winner(self):
        self.play(self.player1)
        self.play(self.player2)
        while self.player1.score==self.player2.score:
            print("Playing again due to draw ..")
            self.play(self.player1)
            self.play(self.player2)
        if self.player1.score>self.player2.score:
            return self.player1 
        else:
            return self.player2


dice = Dice()
coin = Coin()

players = [
    Player("Ansar"),
    Player("Hussnain"),
    Player("Zahid"),
    Player("Ali")
]

print("MATCH 1")
game1 = Game(players[0], players[1], dice, coin)
winner1 = game1.select_winner()
print(winner1.name,"won the game\n\n")

print("MATCH 2")
game2 = Game(players[2], players[3], dice, coin)
winner2 = game2.select_winner()
print(winner2.name,"won the game\n\n\n")

print("FINAL MATCH")
final_game = Game(winner1, winner2, dice, coin)
final_winner = final_game.select_winner()
print(final_winner.name,"won the final game\n_____________________")
```
